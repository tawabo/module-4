```python
import pandas as pd
from pathlib import Path
%matplotlib inline
```

```python
whale_csvpath = Path("../Starter_Code 4/Resources/whale_navs.csv")
```

```python
whales_df = pd.read_csv(whale_csvpath,index_col = 'date', parse_dates=True,infer_datetime_format=True)
whales_df.head()
```

```python
whales_daily_return = whales_df.pct_change().dropna()
whales_daily_return.head()
```

```python
whales_daily_return.plot(figsize = (10,7), title = "Whales to the moon")
```

```python
whales_cumlative_returns = (1 + whales_daily_return).cumprod()
whales_cumlative_returns.tail()
```

```python
whales_cumlative_returns.plot(figsize = (10,7), title = "Whales to the moon")
```
